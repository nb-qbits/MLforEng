```python
import os
import sys
from pathlib import Path

# Move from notebooks/ to project root (MLforEng)
project_root = Path("..").resolve()
os.chdir(project_root)

# Make sure project root is on sys.path so `import mlforeng` works
if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

print("CWD now:", os.getcwd())

from mlforeng.train import TrainConfig, train
```

```python
config = TrainConfig(
    # model_name="logreg",          # you can change to "rf" later
    model_name="rf",
    n_samples=500,
    n_features=10,
    test_size=0.2,
    save_model_name="nb_logreg_test",
)
config
```

```python
results = train(config)
results
```

```python
from mlforeng.predict import load_trained_model, predict_array
from mlforeng.data import load_example_dataset

# Load the model you trained in the notebook earlier
loaded = load_trained_model("nb_logreg_test")
loaded.meta
```

```python
# Build a small test batch with the same number of features
cfg = loaded.meta.get("config", {})
n_features = cfg.get("n_features", 10)

splits = load_example_dataset(
    n_samples=200,
    n_features=n_features,
    test_size=0.2,
)
X_test = splits.X_test[:5]

preds = predict_array(loaded, X_test)
preds
```
