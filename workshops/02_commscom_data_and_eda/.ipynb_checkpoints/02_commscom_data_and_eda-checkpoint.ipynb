# CommsCom – Customer Churn Baseline Modeling

In this module we will:

1. Load the CommsCom customer churn dataset.
2. Prepare a clean binary churn label (0 = stayed, 1 = churned).
3. Split into train/test sets.
4. Build baseline churn models:
   - Logistic Regression
   - Random Forest
5. Evaluate them using classification metrics and ROC–AUC.

This notebook uses the `mlforeng` package you already have in your project.


````python
import os
import sys
from pathlib import Path

import pandas as pd

# Assuming this notebook lives in: MLforEng/workshops/02_commscom_churn_baseline
project_root = Path.cwd().parents[1]
print("Project root:", project_root)

if str(project_root) not in sys.path:
    sys.path.append(str(project_root))

os.chdir(project_root)
print("CWD:", os.getcwd())

from mlforeng.data_churn import (
    load_churn_raw,
    load_churn_dictionary,
    train_test_churn,
)
````

````python
# Load Customer Data - CommsCom customer dataset into memory
df = load_churn_raw()
df.head()
````

````python
# Total info on data: 7043 customers, 38 features, data type etc
df.info()
````

````python
# Check distribution of data
df["Customer Status"].value_counts(normalize=True).to_frame("fraction")
````

````python
# Create lables 0,1
data_dict = load_churn_dictionary()
data_dict.head()
````

````python
# Split data - train/test
splits = train_test_churn(test_size=0.2, stratify=True)

X_train = splits.X_train
X_test = splits.X_test
y_train = splits.y_train
y_test = splits.y_test

X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

````python
# preprocessing pipeline
y_train.value_counts(normalize=True).to_frame("train_fraction")
````

````python
y_test.value_counts(normalize=True).to_frame("test_fraction")
````

````python
X_train.dtypes.value_counts()
````

````python
num_cols = X_train.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_cols = X_train.select_dtypes(include=["object", "bool"]).columns.tolist()

len(num_cols), len(cat_cols), num_cols[:5], cat_cols[:5]
````

````python
# Standardize the data, 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer

# Numeric features: impute missing values, then standardize
numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
    ]
)

# Categorical features: impute missing values, then one-hot encode
categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(
            handle_unknown="ignore",
            sparse_output=False,
        )),
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_cols),
        ("cat", categorical_transformer, cat_cols),
    ]
)

logreg_clf = Pipeline(
    steps=[
        ("preprocess", preprocessor),
        ("model", LogisticRegression(max_iter=1000)),
    ]
)
````

````python
logreg_clf.fit(X_train, y_train)
````

````python
# Train random Forest Model
from sklearn.ensemble import RandomForestClassifier

rf_clf = Pipeline(
    steps=[
        ("preprocess", preprocessor),   # reuse the same preprocessor
        ("model", RandomForestClassifier(
            n_estimators=200,
            random_state=42,
        )),
    ]
)
````

````python
# Test Random Forest on test data
# Compare performance to Logistic Regression

from sklearn.metrics import classification_report, roc_auc_score

y_pred_logreg = logreg_clf.predict(X_test)
y_proba_logreg = logreg_clf.predict_proba(X_test)[:, 1]

print("=== Logistic Regression – Classification Report ===")
print(classification_report(y_test, y_pred_logreg, digits=3))

roc_auc_logreg = roc_auc_score(y_test, y_proba_logreg)
roc_auc_logreg
````

````python
rf_clf.fit(X_train, y_train)
````

````python
y_pred_rf = rf_clf.predict(X_test)
y_proba_rf = rf_clf.predict_proba(X_test)[:, 1]

print("=== Random Forest – Classification Report ===")
print(classification_report(y_test, y_pred_rf, digits=3))

roc_auc_rf = roc_auc_score(y_test, y_proba_rf)
roc_auc_rf
````

````python
roc_auc_logreg, roc_auc_rf
````

````python
from sklearn.metrics import confusion_matrix
import numpy as np

cm_rf = confusion_matrix(y_test, y_pred_rf)
cm_rf
````
